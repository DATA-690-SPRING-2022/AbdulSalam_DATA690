```python
# function for cleaning the data of the data frame

def cleanse_data(data):
    data = str(data).replace("*","").replace(" ","").replace(",","").replace("$","").replace("Billion","000000000")
    if "cents" in data:
        data = float(data.replace("cents","")) / 100
    return int(float(data))
```

## 8. Cleanse the data of the dataframe and convert all the data to "int" type using apply() function

```python
for col in df.columns:
    df[col] = df[col].apply(cleanse_data)
```
