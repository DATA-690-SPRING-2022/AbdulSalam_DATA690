```python
import pandas as pd
import plotly.express as px

DATA_URL = 'https://raw.githubusercontent.com/DATA-690-SPRING-2022/WANG/main/assignment_02/survey.csv'
```

```python
df = pd.read_csv(DATA_URL)
# print(df.head())
# print(df.tail())
df.sample(3)
```

# Step 2: Data Aggregation

```python
df_groupby = df.groupby("2. My primary job function")['Timestamp'].count().reset_index()
df_groupby
```

```python
df_groupby.columns = ["job functions","count"]
df_groupby
```

# Step 3: Data Visualization

```python
fig = px.pie(df_groupby, values='count', names='job functions', title='data 690 class job distributions')
fig.show()
```

```python
fig = px.bar(df_groupby, y='count', x='job functions',color = "job functions", title='data 690 class job distributions')
fig.show()
```
